# Letra - B

```python
OrigCol = dataf.columns.tolist()

def TrocarColuna(df,ListCol, col1, col2):
    aux = ListCol[col1]
    ListCol[col1] = ListCol[col2]
    ListCol[col2] = aux
    
    df = df.reindex(columns = ListCol)
    return df

dataf = TrocarColuna(dataf, OrigCol, 0, 1)
dataf
```
